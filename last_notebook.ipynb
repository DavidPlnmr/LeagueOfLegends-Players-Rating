```python
import numpy as np
import pandas as pd
from ydata_profiling import ProfileReport
# Matplotlib
import matplotlib.pyplot as plt
# Seaborn
import seaborn as sns
```

```python
file = "Worlds_2021_MainEvent.csv"
df = pd.read_csv(file)
```

```python
df.head(10)
```

```python
# Create role columns
roles = ["TOP", "JUNGLE", "MIDDLE", "BOTTOM", "SUPPORT"]
# If the participantId is 1 or 6, the role is TOP
# If the participantId is 2 or 7, the role is JUNGLE
# If the participantId is 3 or 8, the role is MIDDLE
# If the participantId is 4 or 9, the role is BOTTOM
# If the participantId is 5 or 10, the role is SUPPORT
df["role"] = df["participantId"].apply(lambda x: roles[(x - 1) % 5])
```

```python
df.head(10)
```

```python
df.columns.values
```

```python
df = df[[
            'gameId', 
            'participantId', 
            'summonerName', 
            'championId',
            'championName', 
            'teamId', 
            'win',
            'kills',
            'deaths',
            'assists',
            'totalDamageDealtToChampions',
            'goldEarned',
            'totalMinionsKilled',
            'champLevel',
            'wardsPlaced',
            'wardsKilled',
            "totalHeal",
            'patch',
            "damageSelfMitigated",
            "totalDamageTaken",
            "role"
            ]]
```

```python
# Make column for Champion-Role-Patch
df['champ-role-patch'] = df['championName'] + '-' + df['role'] + '-' + str(df['patch'])
```

```python
# Now the goal is to make derived data.
# For example, we will make a new columns which is Kill + Assists
df["k+a"] = df["kills"] + df["assists"]

# See all the columns in the data set
df.columns

# Let's rename the column "totalGold" to "gold"
df.rename(columns={"goldEarned": "gold"}, inplace=True)

# Rename totalHeal to heal
df.rename(columns={"totalHeal": "heal"}, inplace=True)

# Rename totalDamageDealtToChampions to damage
df.rename(columns={"totalDamageDealtToChampions": "damage"}, inplace=True)

# Rename totalDamageTaken to damageTaken
df.rename(columns={"totalDamageTaken": "damageTaken"}, inplace=True)

# Rename totalMinionsKilled to cs
df.rename(columns={"totalMinionsKilled": "cs"}, inplace=True)

df.columns
```

```python
# Deaths
df['total'] = df.groupby(['gameId', 'teamId'])['deaths'].transform('sum')
# If the total is equal to 0, then we set it to 0.000001 to avoid division by 0
df['total'] = df['total'].apply(lambda x: 0.01 if x == 0 else x)

df['deaths'] = df.apply(lambda row: row['deaths'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# k+a
df['total'] = df.groupby(['gameId', 'teamId'])['k+a'].transform('sum')
# Add kills of the game in total column
df['total'] = df.apply(lambda row: row['total'] + df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['kills'].iloc[0], axis=1)

df['k+a'] = df.apply(lambda row: row['k+a'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# damage
df['total'] = df.groupby(['gameId', 'teamId'])['damage'].transform('sum')
df['damage'] = df.apply(lambda row: row['damage'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# gold
df['total'] = df.groupby(['gameId', 'teamId'])['gold'].transform('sum')
df['gold'] = df.apply(lambda row: row['damage'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# cs
df['total'] = df.groupby(['gameId', 'teamId'])['cs'].transform('sum')
df['cs'] = df.apply(lambda row: row['cs'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# champLevel
df['total'] = df.groupby(['gameId', 'teamId'])['champLevel'].transform('sum')
df['champLevel'] = df.apply(lambda row: row['champLevel'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# wardsPlaced
df['total'] = df.groupby(['gameId', 'teamId'])['wardsPlaced'].transform('sum')
df['wardsPlaced'] = df.apply(lambda row: row['wardsPlaced'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# wardsKilled
df['total'] = df.groupby(['gameId', 'teamId'])['wardsKilled'].transform('sum')
df['wardsKilled'] = df.apply(lambda row: row['wardsKilled'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# heal
df['total'] = df.groupby(['gameId', 'teamId'])['heal'].transform('sum')
df['heal'] = df.apply(lambda row: row['heal'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# damageSelfMitigated
df['total'] = df.groupby(['gameId', 'teamId'])['damageSelfMitigated'].transform('sum')
df['damageSelfMitigated'] = df.apply(lambda row: row['damageSelfMitigated'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
# damageTaken
df['total'] = df.groupby(['gameId', 'teamId'])['damageTaken'].transform('sum')
df['damageTaken'] = df.apply(lambda row: row['damageTaken'] / df[(df['gameId'] == row['gameId']) & (df['teamId'] != row['teamId'])]['total'].iloc[0], axis=1)

df = df.drop(columns=['total'])
```

```python
df_without_useless = df.select_dtypes(exclude=['object'])
# Also remove the participantId and gameId columns
df_without_useless = df_without_useless.drop(columns=['participantId', 'gameId', "championId", "teamId", "patch", "kills", "assists"], axis=1)
# Coefficient des attributs du dataset via une corrÃ©lation de Pearson
correlation_matrix = df_without_useless.corr()

# Obtenir les coefficients de corrÃ©lation avec l'attribut cible
coefficient = correlation_matrix['win']

# Trier les coefficients de corrÃ©lation par ordre dÃ©croissant
sorted_coefficient = coefficient.abs().sort_values(ascending=False)

# Afficher le classement des coefficients de corrÃ©lation
print(sorted_coefficient)
print("-"*50)
print(coefficient)
```

```python
# Let's standardize the data
# We will use the StandardScaler from sklearn

from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance

#print(df_without_strings.columns.values)
y = df_without_useless["win"]
X = df_without_useless.drop("win", axis=1)

print(np.isinf(X).any())
# Show the infinite values
inf_loc = np.where(X.values >= np.finfo(np.float64).max)
# Now print the infinite values


scaler = StandardScaler()

# Fit the StandardScaler

X_scaler = scaler.fit_transform(X)

# Create a DataFrame with the scaled data
print(X_scaler)

# Let's create the rating for each row now
# Basically each columns will be multiplied by a coefficient and then summed up
# The coefficients will be the weights of the model
# The sum will be the rating

# Drop the win column
coefficient = coefficient.drop("win", axis=0)
weighted_rating = np.dot(X_scaler, coefficient)

print(weighted_rating)

```

```python
# Now let's add the rating on the dataframe

# Reimport the dataset
final_df = pd.read_csv(file)
final_df["role"] = df["participantId"].apply(lambda x: roles[(x - 1) % 5])
final_df["rating"] = weighted_rating 
```
